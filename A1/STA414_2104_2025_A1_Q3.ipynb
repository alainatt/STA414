## Q3.2
[8pts] For the graph shown above, show using the factorization of the joint probabilities whether $A$ is NOT independent of $G$ given $C$. You may suppose that the domain of the variable $G$ is $\mathbf{G}$.

Hint:  $P(C|A) = \sum_{L \in \mathbf{L}} P(C|A,L) P(L) $.

[Type up your answer below]

**Answer:**
\begin{split}
P(G|A,C)
&=\frac{P(G,A,C)}{P(A,C)}\\
&=\frac{P(G|E,C)P(A)P(C|A,L)}{P(A)P(C|A,L)}\\
&=P(G|E,C)
\end{split}

\begin{split}
P(G|C)
&=\frac{P(G,C)}{P(C)}\\
&=\frac{P(G|E,C)P(C|A,L)}{P(C)}\\
\end{split}


Since
$$P(C|A,L) \neq P(C),$$
then we have:
$$P(G \mid A, C) \neq P(G \mid C).$$

Thus, we've shown that $A$ is NOT independent of $G$ given $C$.
